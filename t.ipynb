```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d = {
    "Alex": [30, 40],
    "Brian": [30,40],
    "Cindy": [0, 80],
    "Daniel": [50, 90],
    "Emma": [100,20 ],
    "Fred": [0, 20]
}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
for i in range(len(df.columns)):
    plt.plot(df.iloc[:,i], alpha=0.7, label=df.columns[i])
plt.legend(df.columns, loc=9, fontsize='small', bbox_to_anchor=[0.5,0])
plt.xlim(0, 8)
```
